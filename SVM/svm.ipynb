```python
import matplotlib.pyplot as plt
import numpy as np

# Define line: w.x + b = 0
w = np.array([1, -1])
b = -1

# Create x range
x_vals = np.linspace(-2, 4, 100)
y_vals = x_vals - 1  # rearranged from x - y - 1 = 0

# Point
x0 = np.array([-3, 0])

# Projection of point onto line
w_norm = w / np.linalg.norm(w)
distance = (np.dot(w, x0) + b) / np.linalg.norm(w)
projection = x0 - distance * w_norm

# Plot
plt.figure()
plt.plot(x_vals, y_vals, label="Decision Boundary (w·x + b = 0)")
plt.scatter(x0[0], x0[1], label="Point")
plt.scatter(projection[0], projection[1], label="Projection on Line")
plt.plot([x0[0], projection[0]], [x0[1], projection[1]], linestyle='--', label="Perpendicular Distance")

# Normal vector
plt.arrow(projection[0], projection[1], w_norm[0], w_norm[1], 
          length_includes_head=True, head_width=0.1, label="Normal Vector w")

plt.legend()
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Distance and Angle Geometry of w·x + b")
plt.axis("equal")
plt.show()
```
