```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.datasets import make_classification

x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
y
```

```python
pd.DataFrame(x)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="linear")
svc.fit(x_train,y_train)
print(svc.coef_)
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print("Classification report:\n",classification_report(y_test,y_pred))
print("Confusion matrix:\n",confusion_matrix(y_test,y_pred))
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
rbf=SVC(kernel="rbf")
rbf.fit(x_train,y_train)
y_pred1=rbf.predict(x_test)
print("Classification report:\n",classification_report(y_test,y_pred1))
print("Confusion matrix:\n",confusion_matrix(y_test,y_pred1))
print("Accuracy:\n",accuracy_score(y_test,y_pred1))
```

```python

pol=SVC(kernel="poly")
pol.fit(x_train,y_train)
y_pred2=pol.predict(x_test)
print("Classification report:\n",classification_report(y_test,y_pred2))
print("Confusion matrix:\n",confusion_matrix(y_test,y_pred2))
print("Accuracy:\n",accuracy_score(y_test,y_pred2))
```

```python

sig=SVC(kernel="sigmoid")
sig.fit(x_train,y_train)
y_pred3=sig.predict(x_test)
print("Classification report:\n",classification_report(y_test,y_pred3))
print("Confusion matrix:\n",confusion_matrix(y_test,y_pred3))
print("Accuracy:\n",accuracy_score(y_test,y_pred3))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={"C":[0.1,1,10,100,1000],
            "gamma":[1,0.1,0.01,.001,0.0001],
            "kernel":['rbf']}

grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
grid.fit(x_train,y_train)
grid.best_params_
```

```python

y_pred4=grid.predict(x_test)
print("Classification report:\n",classification_report(y_test,y_pred4))
print("Confusion matrix:\n",confusion_matrix(y_test,y_pred4))
print("Accuracy:\n",accuracy_score(y_test,y_pred4))
```
